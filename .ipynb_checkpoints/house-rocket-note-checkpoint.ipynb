## 2.1. Hipothesis

### 2.1.1. House level

Garage- Houses with a garage should cost more because it is an extra space in the house and car caring.

Bedrooms - Houses with 2+ bedrooms should have cost more because houses with 2+ bedrooms usually cost more than houses with only 1 bedroom.

Bathrooms - Houses with 2+ bathrooms should have a higher cost because more bathrooms in a house adds value to it. 

Floors - Houses with an Attic, basement or both should have higher cost because another it adds room and value to the house.

Attic - Houses with an Attic should have cost more because it is an extra room or space in the built area of a house and adds more value to it.

Basement - Houses with an Attic should have cost more because it is an extra room or space in the built area of a house and adds more value to it.

Lot area - house with more Lot area should have cost more because more square meters in the land  adds more value to it.

Built area - Houses with more Built area should have cost more because the house becomes bigger and it means more space, bathrooms among other attributes.

Backyard - Houses with backyard should have cost more because this extra space adds more value to the built area of a house.	

Condition - Houses with high conditions should have cost more than house with low condition because a higher  condition adds more value to the house.



### 2.1.2. Customer level

Employment Status - Customers with that are employed should have better purchasing potential because of their jobs and monthly salary

Marital Status - Customers that are married should have a better purchasing potential because if both have jobs the income level should increase.

Income - Customers with income over 80k annual income should have better purchasing potential because they could pay higher mortgages.

Age - Costumers with 30+ should have a better purchasing potential because usually at this age onwards customer should have stable jobs and more mature. 

Credit History - Customers with good credit check should have better money management because they are less risky to do not honor their bills.


### 2.1.3. Location level

Zipcode - Houses located in wealthy zip codes should have cost more because wealthy neighbourhoods increases the value of a house.

Shopping mall - Houses located next to shopping malls should have cost more because shopping malls have a range of basic needs such as groceries, pharmacies among others and entertainment.

CBD - Houses close to the CBD should have cost more because several companies have offices in the CBD and people need to commute to work and living close to the CBD minimizes to commute to work.

Lake - House located next to the lakes should have more cost because of the view and access and entertainment these places deliver.

Local centre - Houses located next to local centres should have cost more because usually local centres has a variety of shops, cafes and groceries next to the houses and become more convenient and easier to local residents go shopping adding more value to the house nearby. 

Public transport - Houses located next to stations should have cost more because of public transport availability next to the houses.



### 2.1.4. Macro Level 

Market - House during economic hardship should have cost less because usually the market drops and houses become cheaper.
